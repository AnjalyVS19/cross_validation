```python
from sklearn.model_selection import KFold,LeaveOneOut,LeavePOut,cross_val_score
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

```python
iris.data
```

```python
iris.target
```

# KFold

```python
kfindex=KFold(n_splits=10,shuffle=True,random_state=10)
```

```python
for train_x,test_x in kfindex.split(iris.data):
    print(train_x)
    print(test_x)
    print()
    print()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
for train_x,test_x in kfindex.split(iris.data):
    xtrain,xtest=iris.data[train_x],iris.data[test_x]
    ytrain,ytest=iris.data[train_x],iris.data[test_x]
    model.fit(xtrain,ytrain)
    print(model.score(xtest,ytest))
```

```python
score=cross_val_score(model,iris.data,iris.target,cv=kfindex)
print(score,score.mean())
```

# LeaveOneOut

```python
loindex=LeaveOneOut()
```

```python
for train_x,test_x in loindex.split(iris.data):
    print(train_x)
    print(test_x)
    print()
    print()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
score=cross_val_score(model,iris.data,iris.target,cv=loindex)
print(score,score.mean())
```
